```python
def spect_summ_th(st, th = 0):
  top = '/content/' + st

  avgs = []
  mins = []
  maxs = []
  stds = []
  lens = [] 

  for pic in  os.listdir(top):
    path = top + '/' + pic
    img = cv2.imread(path, 0)
    img = img[img > th]
    avgs.append(np.average(img))
    mins.append(np.min(img))
    maxs.append(np.max(img))
    stds.append(np.std(img))
    lens.append(len(img))

  aa = np.average(avgs)
  ami = np.average(mins)
  ama = np.average(maxs)
  ast = np.average(stds)
  ale = np.average(lens)

  print("Average {} Minimums {} Maximums {} Std {} Len {}".format(aa,ami,ama,ast, ale))
```

```python
for comm in used_commands:
  print(comm)
  spect_summ_th(comm, th = 35)
```

```python
def get_spect_key_th(st, th = 0):
  top = '/content/' + st
  
  avgs = []
  stds = []
  lens = [] 

  for pic in  os.listdir(top):
    path = top + '/' + pic
    img = cv2.imread(path, 0)
    img = img[img > th]
    avgs.append(np.average(img))
    stds.append(np.std(img))
    lens.append(len(img))

  aa = np.average(avgs)
  ast = np.average(stds)
  ale = np.average(lens)

  return aa, ast, ale
```

```python
min_val = 35

vec_points = {}
vec_points['min_val'] = min_val

for comm in used_commands:
  print(comm)
  print(get_spect_key_th(comm, th = min_val))
  vec_points[comm] = get_spect_key_th(comm, th = min_val)
```

```python
vec_points['min_val']
```

```python
def predict(path, vec_dict, print_nice = False):
  th = vec_dict['min_val']

  img = cv2.imread(path, 0)
  img = img[img > th]
  asvec = [np.average(img),np.std(img),len(img)]
  dist_o = float('inf')
  command = ''
  for com in used_commands:
    dist = np.linalg.norm(asvec - np.array(vec_dict[com]))

    if(dist<dist_o):
      dist_o = dist
      command = com
  
  if print_nice:
    print(used_commands.index(command) ,command, dist_o)

  return used_commands.index(command)
```

```python
predict('/content/up/026290a7_nohash_0.png', vec_points)
```

```python
def get_acc(st,vec_dict, th = 0, n_points = 100):
  top = '/content/' + st
  true_val = used_commands.index(st)

  acc = 0
  counter = 0
  for pic in  os.listdir(top):
    path = top + '/' + pic
    pred = predict(path, vec_dict)
    if pred == true_val:
      acc += 1



  return acc/len(os.listdir(top))
```

```python
get_acc('up',vec_points)
```
